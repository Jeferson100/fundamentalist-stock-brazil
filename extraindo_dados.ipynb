```python
import requests
import pandas as pd

class ObterOpcoes:
    def __init__(self, ativo):
        self.ativo = ativo
        
    def listar_opcoes(self, vencimento):
        url = f'https://opcoes.net.br/listaopcoes/completa?cache=2025-1-9_15h0&au=True&uinhc=1598852945&idLista=ML&idAcao={self.ativo}&listarVencimentos=false&cotacoes=true&vencimentos={vencimento}'
        requests_opcoes = requests.get(url).json()
        opcoes_ativas = [[self.ativo, vencimento, opcao[0].split('_')[0],opcao[2],opcao[3],opcao[4], opcao[5], opcao[8]] for opcao in requests_opcoes['data']['cotacoesOpcoes']]
        return pd.DataFrame(opcoes_ativas, columns=['Ativo', 'Vencimento', 'Codigo', 'Tipo', 'Modelo', 'Itm/Otm/Atm', 'Strike', 'Preco'])
    
    def listar_todos_vencimentos(self):
        url = f'https://opcoes.net.br/listaopcoes/completa?cache=2025-1-9_14h10&au=True&uinhc=1598852945&idLista=ML&idAcao={self.ativo}&listarVencimentos=true&cotacoes=true'
        requests_datas = requests.get(url).json()
        vencimentos = [vencimento['value'] for vencimento in requests_datas['data']['vencimentos']]
        return vencimentos
    
    def listar_opcoes_todos_vencimentos(self):
        vencimentos = self.listar_todos_vencimentos()
        return pd.concat([self.listar_opcoes(vencimento=datas) for datas in vencimentos])
        
        
        
```

```python
opc = ObterOpcoes('ABEV3')
opc_ambv= opc.listar_opcoes_todos_vencimentos()
```
